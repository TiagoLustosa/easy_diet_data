```python
import json
from sklearn.linear_model import LinearRegression

# Input data
with open('C:/projects/easy_diet_data/user_data.json', 'r', encoding='utf-8') as input_file:
         user_data_input = json.load(input_file)
# Output data
with open('C:/projects/easy_diet_data/mealResults.json', 'r', encoding='utf-8') as input_file:
         user_data_output = json.load(input_file)

# Create a Linear Regression model
model = LinearRegression()

# Reshape the input data to a 2D array
X = [[x] for x in user_data_input]

# Fit the model using the input and output data
model.fit(X, user_data_output)

# Predict the output for a new input value
with open('C:/projects/easy_diet_data/new_data_to_regression.json', 'r', encoding='utf-8') as input_file:
         new_input = json.load(input_file)
predicted_output = model.predict(new_input)

# Print the predicted output
print(predicted_output[0])
print('hello')
```

```python
from pyomo.environ import *

# Define the optimization model
model = ConcreteModel()

# Define the decision variables
model.x = Var(['chicken', 'lentils'], domain=NonNegativeReals)

# Define the objective functions
def protein_rule(model):
    return 31*model.x['chicken'] + 9*model.x['lentils']

def calorie_rule(model):
    return 239*model.x['chicken'] + 116*model.x['lentils']

model.protein_obj = Objective(rule=protein_rule, sense=maximize)
model.calorie_obj = Objective(rule=calorie_rule, sense=maximize)

# Define the constraints
def protein_constraint_rule(model):
    return 31*model.x['chicken'] + 9*model.x['lentils'] >= 150

def calorie_constraint_rule(model):
    return 239*model.x['chicken'] + 116*model.x['lentils'] <= 2000

model.protein_constraint = Constraint(rule=protein_constraint_rule)
model.calorie_constraint = Constraint(rule=calorie_constraint_rule)

# Solve the optimization problem
solver = SolverFactory('glpk')
results = solver.solve(model)

# Print the results
print("Chicken quantity: ", value(model.x['chicken']))
print("Lentils quantity: ", value(model.x['lentils']))
```
